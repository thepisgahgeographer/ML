### BMI = Weight/Height squared

```python
weight = float(input("What is your weight"))
height = float(input("What is your height"))

bmi = round(weight / height**2)
if bmi < 18.5:
    print("Underweight")
elif bmi <25:
    print("Normal Weight")
else:
    print("Overweight")

print(bmi)
```

```python
year = int(input("Which year do you want to check?"))

if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print("Leap Year")
    else:
        print("Leap Year")
else:
    print("Not Leap")
```
